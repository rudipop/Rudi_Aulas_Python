#Matplolib e Seaborn

```
# importar as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
x = list(range(1,51))# gerar numeros de 1 a 50, indices
y =np.random.normal(size=50)
print(x)
print(y)
```

```
# plotar série histórica, x indice e y valor associado
plt.plot(x, y, color = 'green', marker='v') #; ponto e virgula suprime as mensagem no gráfico
plt.show()
```

```
plt.plot(x, y, 'bo') #; ponto e virgula suprime as mensagem no gráfico
plt.show()
```

```
plt.plot(x, y, 'r+-') #; ponto e virgula suprime as mensagem no gráfico
plt.show()
```

```
# grafico de dispersão
plt.scatter(x,y, color='green');
```

```
plt.figure(figsize=(20,5)) # define o tamanho do grafico, usualmente fazer a cada gráfico, o padrão 8x5? só o gráfico a seguir
plt.plot(x, y, 'r+-') #; ponto e virgula suprime as mensagem no gráfico
plt.show()
```

```
# boxplot
plt.boxplot(y, showmeans = True);# compreende os dados que não são outliners "muito distantes" 25/50/25, linha laranja é a mediana
```

```
y_outlier = np.concatenate((y,[4])) # gerar outlier parar teste, concatenando com 4 o valor extrapola
plt.boxplot(y_outlier);
```

```
# violin mostra também a distribuição dos dados
plt.violinplot(y);
```

```
plt.violinplot(y_outlier);
```

```
# violin usando seaborn
sns.violinplot(data=y, color='0.8')
sns.stripplot(data=y);
```

```
#histogramas
plt.hist(y);# padrão
```

```
plt.hist(y, bins=20, color='red');
```

```
y2 = np.random.normal(size=50)# mostrando os dois graficos sobrepostos
plt.hist(y, alpha=0.7)# o alpha adiciona certa transparência e pode ver os doi graficos
plt.hist(y2, alpha=0.7)
plt.show()
```

```
plt.hist(y, alpha=0.7, label='y')# o alpha adiciona certa transparência e pode ver os doi graficos
plt.hist(y2, alpha=0.7, label='y2')
# colocar titulos, legenda,etc
plt.title('histograma de y e y2')
plt.xlabel('nome do eixo x')
plt.ylabel('nome do eixo y')
plt.legend()
plt.savefig('histograma.png')# salva o gráfico
plt.show()
```

```
#boxplot
plt.boxplot([y, y2], labels=['y','y2'])# um atributo, idade e renda por exemplo
plt.ylabel('valor')
plt.title('Meus boxplots')
plt.show
```

```
from sklearn.datasets import load_iris
data = load_iris()
x = data.data
y = data.target

# juntar atributos com especies

base = np.column_stack((x,y))
base.shape
```

```
import pandas as pd
df = pd.DataFrame(base)
df.head()
```

```
cols = data.feature_names
cols.append('Species')
cols
```

```
df.columns = cols
df.head()
```

```
# verificar correlação entre os dados

correlations = df.corr()
correlations
```

```
# mapa de calor seaborn
sns.heatmap(correlations, vmin = -1, vmax = 1);
```
