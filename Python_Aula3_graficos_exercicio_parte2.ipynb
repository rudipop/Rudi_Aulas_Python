```
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-bright')
import seaborn as sns

```

```
TendMundi = pd.read_csv('TendenciasMundiais.csv')
TendMundi
```

```
Ferti1960 = TendMundi['FertilityRate.1960']
Ferti2013 = TendMundi['FertilityRate.2013']
Vida1960 = TendMundi['LifeExp.1960']
Vida2013 = TendMundi['LifeExp.2013']
Pais = TendMundi['CountryName']





```

```
TendMundi.describe()
```

```
plt.figure(figsize=(20,5))
plt.plot(Pais,Ferti1960)
plt.hist(Ferti1960, alpha=0.7)
plt.hist(Vida1960, alpha=0.7)
plt.hist(Ferti2013, alpha=0.7)
plt.hist(Vida2013, alpha=0.7)
plt.title('RelaÃ§Ã£o Expectativa de Vida e Taxa de Natalidade')
plt.xlabel('nome do eixo x')
plt.ylabel('nome do eixo y')
plt.legend()
plt.savefig('histograma.png')# salva o grÃ¡fico
plt.show()
```

```
plt.figure(figsize=(15,8))
plt.scatter(TendMundi['FertilityRate.1960'], TendMundi['LifeExp.1960'], alpha=0.7, label='1960')
plt.scatter(TendMundi['FertilityRate.2013'], TendMundi['LifeExp.2013'], alpha=0.7, label='2013')
plt.legend()
plt.title('Fertilidade x Expectativa de Vida')
plt.xlabel('Fertilidade')
plt.ylabel('Expectativa de Vida')
plt.legend()
plt.savefig('dispersao.png')# salva o grÃ¡fico
plt.show()                           
```

```
plt.figure(figsize=(20,20))
plt.scatter(TendMundi['FertilityRate.1960'], TendMundi['LifeExp.1960'], alpha=0.7, label='1960')
plt.scatter(TendMundi['FertilityRate.2013'], TendMundi['LifeExp.2013'], alpha=0.7, label='2013')
plt.legend()
plt.title('Fertilidade x Expectativa de Vida')
plt.xlabel('Fertilidade')
plt.ylabel('Expectativa de Vida')
plt.legend()

# colocar os codigos dos paises
for i, cc in enumerate(TendMundi['CountryCode']):
    plt.annotate(cc, (TendMundi['FertilityRate.1960'][i], TendMundi['LifeExp.1960'][i]))
    plt.annotate(cc, (TendMundi['FertilityRate.2013'][i], TendMundi['LifeExp.2013'][i]))

plt.show()         
```

```
plt.figure(figsize=(20,20))
plt.scatter(TendMundi['FertilityRate.1960'], TendMundi['LifeExp.1960'], alpha=0.7, label='1960')
plt.scatter(TendMundi['FertilityRate.2013'], TendMundi['LifeExp.2013'], alpha=0.7, label='2013')
plt.legend()
plt.title('Fertilidade x Expectativa de Vida')
plt.xlabel('Fertilidade')
plt.ylabel('Expectativa de Vida')
plt.legend()

# colocar os codigos dos paises
for i, cc in enumerate(TendMundi['CountryCode']):
  if cc == 'BRA': # condiional para localizar o Brasil no grÃ¡fico
    plt.annotate(cc, (TendMundi['FertilityRate.1960'][i], TendMundi['LifeExp.1960'][i]))
    plt.annotate(cc, (TendMundi['FertilityRate.2013'][i], TendMundi['LifeExp.2013'][i]))

plt.show()         
```
